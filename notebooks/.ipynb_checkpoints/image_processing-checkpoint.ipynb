## Image Processing
1. load the images
2. convert them to Numpy Arrays
3. resize them 

```python
def read_npArray_resize_image(image_path, x_pixel, y_pixel):
    """
    This function read '.jpg' image using  Python Pillow library (PIL)
    then resize the image using 
    INPUTS: 
    - image_path: 
        full path of the image including the image name
        e.g., '../team-ASL-github-testing-dataset/asl_alphabet_train/asl_alphabet_train/A/A2292.jpg'
    - x_pixel and y_pixel: 
        to resize the image if the original size looks unnecessarily
        e.g., 60x60 if we want to resize them from 200x200
    """
    return skimage.transform.resize(np.asarray(Image.open(image_path)), (x_pixel, y_pixel, 1))
```

# Convolutional Neural Networks

Actetacture Details:
- 2 Hidden Layers (128, 64)
- 1 Conv2D with 6 filters and 3x3 kernel size
- MaxPooling2D
- 

# |||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||

```python
model = Sequential()

# Add a convolutional layer.

model.add(Conv2D(filters = 6,       #number of filters to use
              kernel_size = (3, 3),   #dimensions of the filters
              activation = 'relu',    #activation function
              input_shape = (x_pixel, y_pixel, 1)))  #shape of image

# dimensions of region of pooling

# cnn.add(MaxPooling2D(pool_size = (2, 2)))   #filter dimensions
# cnn.add(Conv2D(16, kernel_size = (3, 3), activation = 'relu'))
# cnn.add(MaxPooling2D(pool_size = (2, 2)))



model.add(Flatten(input_shape = (x_pixel, y_pixel, 1))) # INPUT LAYER

model.add(Dense(128*3, activation = 'relu'))
model.add(Dense(64*2, activation = 'relu'))
model.add(Dense(29, activation = 'softmax'))

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

# Fit model on training data.
history = model.fit(X_train, y_train_cat, validation_data = (X_test, y_test_cat),
                   batch_size = 512*2, # 60000/256 = 235 weight updates per epoch
                   epochs = 500,
                   verbose = 1) 
```

```python
# Check out our train loss and test loss over epochs.
train_loss = history.history['accuracy']
test_loss = history.history['val_accuracy']

score = model.evaluate(X_test,y_test_cat, verbose=0)
print('Test set Score [Loss, Accuracy]=',score)
score = model.evaluate(X_train,y_train_cat, verbose=0)
print('Train Set Score [Loss, Accuracy]=',score)


# Set figure size.
plt.figure(figsize=(12, 8))

# Generate line plot of training, testing loss over epochs.
plt.plot(train_loss, label='Training Loss', color='#185fad')
plt.plot(test_loss, label='Testing Loss', color='orange')

# Set title
plt.title('Training and Testing Loss by Epoch', fontsize = 25)
plt.xlabel('Epoch', fontsize = 18)
plt.ylabel('Categorical Crossentropy', fontsize = 18)
# plt.xticks(np.arange(10), np.arange(10))

plt.legend(fontsize = 18);
```
